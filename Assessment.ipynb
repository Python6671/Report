```python
# Define number of trials and success probability
n_trials = 10  # e.g., number of customers in a party
p_success = 0.5  # Assume a 50% chance of tipping above a threshold

# Generate binomial distribution data
binomial_data = np.random.binomial(n_trials, p_success, 1000)

# Plot
plt.figure(figsize=(10, 5))
sns.histplot(binomial_data, bins=range(n_trials + 2), kde=False, stat='probability')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successful Tips (Above Threshold)')
plt.ylabel('Probability')
plt.grid()
plt.show()
```

```python
# Average number of customers (λ) per hour based on the size of parties
lambda_rate = tips['size'].mean()

# Generate Poisson distribution data
poisson_data = np.random.poisson(lambda_rate, 1000)

# Plot
plt.figure(figsize=(10, 5))
sns.histplot(poisson_data, bins=range(0, 10), kde=False, stat='probability')
plt.title('Poisson Distribution (λ = Average Group Size)')
plt.xlabel('Number of Customers')
plt.ylabel('Probability')
plt.grid()
plt.show()
```
